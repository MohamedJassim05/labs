##### 1.❓Are there better or worse days/times in terms of sales? Plot a chart to show the result

```python
# Extracting dates from transaction date

coffee_df['transaction_date'].dt.date
```

```python
# Creating time categories function to return Morning, Noon, Evenging, Night based on transaction_time


def time_category(time):

    hour = int(str(time).split(':')[0])

    if hour >=5 and hour < 11:
        return 'Morning'
    elif hour >= 12 and hour < 17:
        return 'Noon'
    elif  hour >= 17 and hour < 20:
        return 'Evening'
    else:
        return 'Night'




# use apply to apply the time category into a new column called time_category


coffee_df['time_category'] = coffee_df['transaction_time'].apply(time_category)
coffee_df.head()
```

```python
# Are there better or worse days/times in terms of sales?
coffee_df.groupby(['transaction_day'])['line_item_amount'].sum().sort_values(ascending=False).plot(kind='bar', title='Total Sales by day',rot=90);
# Write your findings
```

> 👉 **Write Your Findings Here**: 
-- -Most of the days appear to have similar numbers in terms of sales , while monday stands out to be the highest day of sales considering its the first day after the weekend in US.

```python
# Are there better or worse days/times in terms of sales?
coffee_df.groupby(['time_category'])['line_item_amount'].sum().sort_values(ascending=False).plot(kind='bar', title='Total Sales by time',rot=90);

# Write your findings

```
